## Evaluate

```python
import numpy as np
from sklearn.metrics import roc_auc_score

def _ROC_AUC_(rr, pred_rr, p=100):
    """ROC_AUC
    rr: true ranking, 
    pred_rr: predicted ranking, 
    p: total number of variables
    """
    all_vars = np.zeros(p)
    all_vars[rr] = 1
    _rank = np.argsort(pred_rr)
    return 1-roc_auc_score(all_vars, _rank)
```

```python
# reliable mi oracle

# rr_vars is the relevant variables true rank
reliable_mi_score = _ROC_AUC_(rr_vars, self_obj.score_feature_rank())
reliable_mi_order = _ROC_AUC_(rr_vars, self_obj.order_feature_rank())

multi_mi_score = _ROC_AUC_(rr_vars, self_obj1.score_feature_rank())
multi_mi_order = _ROC_AUC_(rr_vars, self_obj1.order_feature_rank())

single_mi_score = _ROC_AUC_(rr_vars, self_obj2.score_feature_rank())
single_mi_order = _ROC_AUC_(rr_vars, self_obj2.order_feature_rank())

single_p_score = _ROC_AUC_(rr_vars, self_obj3.score_feature_rank())
single_p_order = _ROC_AUC_(rr_vars, self_obj3.order_feature_rank())

multi_p_score = _ROC_AUC_(rr_vars, self_obj4.score_feature_rank())
multi_p_order = _ROC_AUC_(rr_vars, self_obj4.order_feature_rank())
```
