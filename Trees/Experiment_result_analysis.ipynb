```python
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
import numpy as np

from plotting import plotDf
from data_generator import dataGenerator2d
```

## EXP1

```python
x, y = dataGenerator2d(n=10000, irr=0, noise=0.05).fit() # noise=0.3
colors = np.array(['#377eb8', '#ff7f00'])
plt.scatter(x['X0'], x['X1'], s=10, color=colors[y])
plt.xlabel('$X_0 \sim cos(X)$')
plt.ylabel('$X_1 \sim sin(X)$')
plt.title('Pure Circle $X \sim U(0, 2\pi)$')
plt.savefig('Exp1_example' + '.pdf', bbox_inches='tight', pad_inches=0)
```

```python
exp1path = 'output/Experiment1/'
onlyfiles = [f for f in listdir(exp1path) if isfile(join(exp1path, f)) and f != '.DS_Store']
onlyfiles
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, sharey=True, sharex=True, figsize=(16,5))
plotDf(onlyfiles, exp1path, ax1, ax2, ax3, 'Experiment1')
```

# Exp4

```python
x, y = dataGenerator2d(n=10000, irr=0, types='interaction').fit()
colors = np.array(['#377eb8', '#ff7f00'])
plt.scatter(x['X0'], x['X2'], s=10, color=colors[y])
plt.xlabel('X3')
plt.ylabel('X1')
plt.title('$Y=Be(\sigma(X_1 -X_2 + X_3))$\n$X_3=X_2X_1$')
plt.savefig('Exp4_example1' + '.pdf', bbox_inches='tight', pad_inches=0)
```

```python
x, y = dataGenerator2d(n=10000, irr=0, types='interaction').fit()
colors = np.array(['#377eb8', '#ff7f00'])
plt.scatter(x['X1'], x['X2'], s=10, color=colors[y])
plt.xlabel('X3')
plt.ylabel('X1')
plt.title('$Y=Be(\sigma(X_1 -X_2 + X_3))$\n$X_3=X_1X_2$')
plt.savefig('Exp4_example2' + '.pdf', bbox_inches='tight', pad_inches=0)
```

```python
exp4path = 'output/Experiment4/'
exp4files = [f for f in listdir(exp4path) if isfile(join(exp4path, f)) and f != '.DS_Store']
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, sharey=True, sharex=True, figsize=(16,5))
plotDf(exp4files, exp4path, ax1, ax2, ax3, 'Experiment4', legend=True,title=True, xalbel_name=True,fig=fig)
plt.savefig('Experiment4.pdf', bbox_inches='tight', pad_inches=0)
```

```python
fig, axs = plt.subplots(4,3, sharey=True, sharex=True, figsize=(16,12))

plotDf(onlyfiles, exp1path, axs[0, 0], axs[0, 1], axs[0, 2], 'Experiment1', legend=True, title=True)
plotDf(exp2files, exp2path, axs[1, 0], axs[1, 1], axs[1, 2], 'Experiment2')
plotDf(exp3files, exp3path, axs[2, 0], axs[2, 1], axs[2, 2], 'Experiment3')
plotDf(exp4files, exp4path, axs[3, 0], axs[3, 1], axs[3, 2], 'Experiment4')
axs[0, 0].set_ylabel('Exp1')
axs[1, 0].set_ylabel('Exp2')
axs[2, 0].set_ylabel('Exp3')
axs[3, 0].set_ylabel('Exp4')
plt.savefig('ExperimentCombine' + '.pdf', bbox_inches='tight', pad_inches=0)
```
